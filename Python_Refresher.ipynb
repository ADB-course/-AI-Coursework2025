```python
# Program to convert days to seconds
def days_to_seconds():
    """
    This function asks the user for number of days,
    converts it to seconds, and displays the result.
    """
    # Ask user for input
    days = float(input("Enter the number of days: "))

    # Calculate seconds (1 day = 24 hours, 1 hour = 60 minutes, 1 minute = 60 seconds)
    seconds = days * 24 * 60 * 60

    # Display the result
    print(f"{days} days is equal to {seconds:,} seconds")

# Call the function
days_to_seconds()



# Program to calculate volume of a sphere
import math

def sphere_volume():
    """
    This function calculates the volume of a sphere
    given its radius using the formula (4/3)πr³.
    """
    # Ask user for radius input
    radius = float(input("Enter the radius of the sphere: "))

    # Calculate volume using exponential operator ** for r³
    volume = (4/3) * math.pi * (radius ** 3)

    # Display the result with 2 decimal places
    print(f"The volume of a sphere with radius {radius} is {volume:.2f}")

# Call the function
sphere_volume()



# Program to calculate area and perimeter of a square using functions

def calculate_area(side_length):
    """
    Calculates the area of a square.
    Formula: area = side × side
    """
    return side_length * side_length

def calculate_perimeter(side_length):
    """
    Calculates the perimeter of a square.
    Formula: perimeter = 4 × side
    """
    return 4 * side_length

def square_calculator():
    """
    Main function that gets user input and displays
    the area and perimeter of a square.
    """
    # Get side length from user
    side = float(input("Enter the side length of the square: "))

    # Calculate area and perimeter using the functions
    area = calculate_area(side)
    perimeter = calculate_perimeter(side)

    # Display results
    print(f"\nSquare with side length {side}:")
    print(f"Area: {area}")
    print(f"Perimeter: {perimeter}")

# Call the main function
square_calculator()



# Program to determine if a character is uppercase or lowercase

def check_case(character):
    """
    Checks if a character is uppercase or lowercase.
    Returns 'Uppercase', 'Lowercase', or 'Not a letter'
    """
    if len(character) != 1:
        return "Please enter exactly one character"

    if character.isupper():
        return "Uppercase"
    elif character.islower():
        return "Lowercase"
    else:
        return "Not a letter"

def case_detector():
    """
    Main function that gets user input and determines
    the case of the character.
    """
    # Get character from user
    char = input("Enter a single character: ")

    # Check the case using the function
    result = check_case(char)

    # Display the result
    print(f"The character '{char}' is: {result}")

# Call the main function
case_detector()


# Implementation of the given pseudocode

def pseudocode_implementation():
    """
    This program implements the given pseudocode:
    BEGIN
    SET x TO 0, y TO 20
    REPEAT
    SUBTRACT 4 FROM y
    ADD 2/y TO x
    UNTIL
    y IS LESS THAN 6
    DISPLAY x
    END
    """
    # Initialize variables as per pseudocode
    x = 0
    y = 20

    # a while loop to implement the REPEAT-UNTIL logic
    # REPEAT-UNTIL loops run at least once and continue until condition is met
    while True:
        # SUBTRACT 4 FROM y
        y -= 4

        # ADD 2/y TO x
        x += 2 / y

        # UNTIL y IS LESS THAN 6
        if y < 6:
            break

    # DISPLAY x
    print(f"The final value of x is: {x}")

# Call the function
pseudocode_implementation()



# Program to input 5 values into an array and calculate their average

def array_average_calculator():
    """
    This program:
    • Uses a loop for a user to continually input 5 values to populate an array
    • Calculates and displays the average of the values input into the array
    """
    # Initialize an empty list (array) to store values
    numbers = []

    print("Please enter 5 numbers:")

    # a for loop to get exactly 5 inputs from the user
    for i in range(5):
        while True:
            try:
                # Get input from user and convert to float
                value = float(input(f"Enter value {i + 1}: "))
                numbers.append(value)
                break  # Exit the while loop if input is valid
            except ValueError:
                print("Invalid input! Please enter a numeric value.")

    # Calculate the average of the values in the array
    # Sum all elements and divide by the number of elements
    average = sum(numbers) / len(numbers)

    # Display the results
    print(f"\nThe values you entered are: {numbers}")
    print(f"The average of these values is: {average:.2f}")

```
