```python
fatality_locations = df[['Date', 'Location', 'Aboard', 'Fatalities']].copy()
fatality_locations.head()
```

```python
max_fatalities = fatality_locations['Fatalities'].max()
deadliest_date = fatality_locations.loc[fatality_locations['Fatalities'] == max_fatalities, 'Date'].values[0]

print(f"The highest number of fatalities occurred on {deadliest_date} with {max_fatalities} fatalities.")
```

```python
# Compare aboard vs fatalities
fatality_locations['Survivors'] = fatality_locations['Aboard'] - fatality_locations['Fatalities']


no_fatalities = fatality_locations[fatality_locations['Fatalities'] == 0]
print(f"There are {len(no_fatalities)} crashes with no fatalities.")
no_fatalities.head()
```

```python
# Split Location column
split_location = fatality_locations['Location'].str.split(',', n=1, expand=True)
fatality_locations['Region'] = split_location[0]
fatality_locations['State/Country'] = split_location[1].str.strip()

fatality_locations.head()
```

```python
top_100 = fatality_locations.sort_values(by='Fatalities', ascending=False).head(100)
top_100.head()
```

```python
# Get top 25 locations by total fatalities
top_25 = (fatality_locations.groupby('State/Country')['Fatalities']
           .sum()
           .sort_values(ascending=False)
           .head(25))

# Create pie chart
plt.figure(figsize=(10,10))
top_25.plot.pie(autopct='%1.1f%%', startangle=90, cmap='tab20')
plt.title('Top 25 Fatalities by Country/State')
plt.ylabel('')
plt.show()
```
